```python
from collections import Counter

allDifferences = []
lastN = None
for n in sopr.recurse().notes:
    if 'Chord' in n.classes:
        continue
    if lastN is not None:
        thisMidi = n.pitch.midi
        lastMidi = lastN.pitch.midi
        allDifferences.append(thisMidi - lastMidi)
    lastN = n
```

```python
count = Counter(allDifferences)
print(count)
```

```python
count.most_common(1)
```
