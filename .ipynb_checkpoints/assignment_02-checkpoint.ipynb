## Working with Corpora

A corpus of pieces is simply a repertory or list of files that the same or similar operations can be performed on each and the results can be aggregated in certain ways.

Generally when you think of corpora think of `for`-loops.  For instance we can think of the four chorales from Assignment 1 as a sort of mini corpus:


```python
miniCorpus = ['bwv66.6', 'bwv10.7', 'bwv255', 'bwv366']

for work in miniCorpus:
    chorale = corpus.parse('bach/' + work)
    numberOfNotes = len(chorale.recurse().notes)
    print(numberOfNotes)
```

In general, you'll want to store this information somewhere to work on it.  And that means setting up a storage unit before iterating (that is, before running the for-loop).  Here we'll create a list to store `notesPerPiece`.

```python
notesPerPiece = []

for work in miniCorpus:
    chorale = corpus.parse('bach/' + work)
    numberOfNotes = len(chorale.recurse().notes)
    notesPerPiece.append(numberOfNotes)
```

Now we can figure out the average number of notes per piece:

```python
import statistics

print(statistics.mean(notesPerPiece))
```

(The number of notes is just a demostration because it's easy to calculate; it's probably not that interesting.  Though perhaps plotting a correlation between the number of notes and the year Bach composed a piece would be interesting).

I want to move on from Bach very soon, but it's worth noting that `music21` comes with a special way of using the Bach chorales as a corpus and to do something with each chorale, called `corpus.chorales.Iterator()`.  Here we'll use Carol Krumhansl's famous mathod for analyzing keys to see what keys chorales are in.  If you run this on your own, just note that the first run might take as long as ten minutes.  After running once, subsequent runs are faster.

```python
keys = []

for chorale in corpus.chorales.Iterator():
    # chorale is already parsed
    if len(chorale.parts) != 4:
        # some chorales have instrumental parts also.  Let's keep it simple.
        continue
    k = chorale.analyze('key.krumhansl')
    # k is a key.Key object.
    # look at https://web.mit.edu/music21/doc/moduleReference/moduleKey.html
    # for details on what you can do with a key
    keys.append(k.tonicPitchNameWithCase)
    
from collections import Counter
print(Counter(keys))
```

So if Krumhansl's algorithm is accurate, a minor, G major, and and g minor are the most common keys for Bach's chorales, with C major less common than either G, A, F, or D major!  (Analyzing `'key'` without `.krumhansl` uses the related algorithm by Bret Arden which puts B-flat major as more common than C also).

## Back to working with files...

Okay, we've done the important step of seeing how we might work with a collection of files, but we still will want to see what we could do with any particular file.  So let's start with one file -- "Jeannie with the light brown hair" by Stephen Foster which we can get via:

```python
jeannie = corpus.parse('foster')
jeannie.show()
```

This is a leadsheet that has `ChordSymbol` objects throughout it.  So one of the things we can do is see which of the notes in the piece are in the prevailing chord.  Let's color the non-harmonic tones red.  We'll use `jeannie.recurse().getElementsByClass('Note')` to get simple notes in the piece, but then use `.getContextByClass('ChordSymbol')` to figure out what chord is playing at the time.  First let's just make sure it works by printing the first 25 notes.

```python
for n in jeannie.recurse().getElementsByClass('Note')[:25]:
    chord_context = n.getContextByClass('ChordSymbol')
    print(n, chord_context)
```

(This sheet isn't entirely properly encoded in that the `N.C.` symbols are just text directions, not proper "NoChord" chord symbols -- this will skew our results ever so slightly with those non-harmonic Ds, but otherwise this should work great).

So now we can see if the pitch of the note is in the chord context:

```python
for n in jeannie.recurse().getElementsByClass('Note'):
    chord_context = n.getContextByClass('ChordSymbol')
    if chord_context is None or 'NoChord' in chord_context.classes:
        continue
    chord_pitch_names = [p.name for p in chord_context.pitches]
    if n.pitch.name not in chord_pitch_names:
        n.style.color = 'red'
        print(f'{n} is not in {chord_context}')
```

Now let's look at our score:

```python
jeannie.show()
```

We can check what beat each of the non-harmonic tones is on, with our counter mechanism:

```python
nonharmonicBeats = []

for n in jeannie.recurse().getElementsByClass('Note'):
    chord_context = n.getContextByClass('ChordSymbol')
    if chord_context is None or 'NoChord' in chord_context.classes:
        continue
    chord_pitch_names = [p.name for p in chord_context.pitches]
    if n.pitch.name not in chord_pitch_names:
        n_beat = n.beat
        nonharmonicBeats.append(n_beat)
        
print(Counter(nonharmonicBeats))
```

As expected, most non-harmonic tones appear on beats 2 and 4, and only one appears on the downbeat (Does this have anything to do with the text "floating" at that moment?)  Let's print the context for each non-harmonic tone:

```python
for n in jeannie.recurse().getElementsByClass('Note'):
    chord_context = n.getContextByClass('ChordSymbol')
    if chord_context is None or 'NoChord' in chord_context.classes:
        continue
    chord_pitch_names = [p.name for p in chord_context.pitches]
    if n.pitch.name not in chord_pitch_names:
        previous_note = n.previous('Note')
        next_note = n.next('Note')
        print(previous_note.nameWithOctave, n.nameWithOctave, next_note.nameWithOctave)
```

## Other things to do with chords

Our two examples already had the chords specified, but there are times when we'll want the system to automatically identify chords for us.  One primative but powerful way to do so is to simply "`chordify`" the whole score, thereby reducing all chords to a single staff:

```python
b = corpus.parse('bach/bwv66.6')
b.chordify().show()
```

I've made a pretty good example of how to use Chordify as Chapter 9 in the music21 User's Guide, so best to read chapters 9 (and probably 10) first before moving on:  https://web.mit.edu/music21/doc/usersGuide/usersGuide_09_chordify.html

Let's use chordify to example the succession of pitchClass cardinalities in the 6th movement of Schoenberg's Opus 19 piano pieces, found in the Corpus:

```python
sch = corpus.parse('schoenberg/opus19/movement6')
sch.show()
```

```python
chordified = sch.chordify()
lastCardinality = 0
for c in chordified.recurse().getElementsByClass('Chord'):
    c.closedPosition(forceOctave=4, inPlace=True)
    cardinality = c.pitchClassCardinality
    if cardinality != lastCardinality:
        c.lyric = str(cardinality)
        lastCardinality = cardinality
chordified.show()
```

As it does in listening to the piece, the isolated solo-melody section jumps out in this analysis.  However, I never noticed until now that the middle section is the only place that uses more than 3 but fewer than 6 distinct pitch classes at a time.
