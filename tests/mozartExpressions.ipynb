# Mozart Expressions

Let's find all the expression marks in Mozart!

```python
from music21 import *
allMozart = corpus.search('mozart')
allMozart
```

```python
for m in allMozart:
    m_parsed = m.parse()
    for n in m_parsed.recurse().notes:
        if n.expressions:
            print(m, n.expressions, n.measureNumber)
```
