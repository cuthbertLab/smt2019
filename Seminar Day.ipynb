```python
for bachPiece in corpus.chorales.Iterator():
    numRepeatedNotes = Counter()
    for partNumber in range(4):
        vp = bachPiece.parts[partNumber]
        lastPS = 9999
        for n in vp.recurse().notes:
            thisPS = n.pitch.ps
            if thisPS == lastPS:
                numRepeatedNotes[partNumber] += 1
            lastPS = thisPS
    partNumberWithMostRepetitions = numRepeatedNotes.most_common(1)[0][0]
    print(partNumberWithMostRepetitions, bachPiece.metadata.title)
```
